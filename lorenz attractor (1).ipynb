```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#parameters
r = 28.0
sig = 10.0
beta = 8.0/3.0

dt = 0.01 #time step
n = 3000  #time = n*dt

x = np.zeros(shape=(n))
y = np.zeros(shape=(n))
z = np.zeros(shape=(n))

#initialise states
x[0] = 1.0
y[0] = 1.0
z[0] = 1.0

for i in range(1,n):
    #differential equations to get change of x for every time step
    dx = sig * (y[i-1] - x[i-1]) * dt
    dy = ((x[i-1] * (r - z[i-1])) - y[i-1]) * dt
    dz = ((x[i-1] * y[i-1]) - (beta * z[i-1])) * dt
    
    #update x y z
    x[i] = x[i-1] + dx
    y[i] = y[i-1] + dy 
    z[i] = z[i-1] + dz
    
    #print(x[i],y[i],z[i])

plt.figure(1)
ax = plt.axes(projection='3d')
#ax.scatter3D(x, y, z,s = 0.01)
ax.plot(x,y,z)
plt.savefig()
```
